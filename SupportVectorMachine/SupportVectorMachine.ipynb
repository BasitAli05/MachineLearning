```python
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
iris = load_iris()
```

```python
iris
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
df0 = df[:50]
df1 = df[50:100]
df2 = df[100:]
```

```python
plt.xlabel("sepal length")
plt.ylabel("sepal width")
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],color='blue',marker='+')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color='green',marker='+')
plt.scatter(df2['sepal length (cm)'],df2['sepal width (cm)'],color='red',marker='+')
```

```python
plt.xlabel("petal length")
plt.ylabel("petal width")
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],color='blue',marker='+')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color='green',marker='+')
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'],color='red',marker='+')
```

```python
from sklearn.model_selection import train_test_split
X = df.drop(['target','flower_name'],axis='columns')
Y = df.target
```

```python
print('--------Inputs-------')
print(X.head())
print('--------Output-------')
print(Y.head())
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```

```python
model.predict([[7,3,6,1.5]])
```

Change values Regularization , Gamma and kernel by trial and error for higher accuracy


```python
model_c = SVC(C=100)
model_c.fit(X_train,Y_train)
model_c.score(X_test,Y_test)
```

```python
model_g = SVC(gamma=4)
model_g.fit(X_train,Y_train)
model_g.score(X_test,Y_test)
```

```python
model_linear_kernel = SVC(kernel='linear')
model_linear_kernel.fit(X_train,Y_train)
model_linear_kernel.score(X_test,Y_test)
```

Train SVM classifier using sklearn digits dataset (i.e. from sklearn.datasets import load_digits) and then,

1. Measure accuracy of your model using different kernels such as rbf and linear.
2. Tune your model further using regularization and gamma parameters and try to come up with highest accurancy score
3. Use 80% of samples as training data size

```python
from sklearn.datasets import load_digits
digits = load_digits()
digits
```

```python
digits.target
```

```python
df = pd.DataFrame(digits.data)
df.head()
```

```python
df['target'] = digits.target
df.head(20)
```

```python
from sklearn.model_selection import train_test_split
X = df.drop('target',axis='columns')
Y = df.target
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
from sklearn.svm import SVC
rbf_model = SVC(kernel='rbf')
rbf_model.fit(X_train, y_train)
rbf_model.score(X_test,y_test)
```

```python
linear_model = SVC(kernel='linear')
linear_model.fit(X_train,y_train)
linear_model.score(X_test,y_test)
```

Regularization and Gamma functions to increase accuracy

```python
model_c = SVC(C=10)
model_c.fit(X_train, y_train)
model_c.score(X_test,y_test)
```

```python
model_g = SVC(gamma=1000)
model_g.fit(X_train, y_train)
model_g.score(X_test,y_test)
```
