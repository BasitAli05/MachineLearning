```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
%matplotlib inline
import math
from word2number import w2n
from num2words import num2words
```

```python
df = pd.read_csv('hiring.csv')
df
```

```python
def convert_words_to_numbers(word):
    try:
        number = w2n.word_to_num(word)
        return number
    except ValueError:
        return word
```

```python
df['experience'] = df['experience'].apply(convert_words_to_numbers)
# m = converted_experience.median()
# mean_word = num2words(m)
df['experience'] = df.experience.fillna(math.floor(df['experience'].mean()))
df
```

```python
m = math.floor(df["test_score(out of 10)"].mean())
df["test_score(out of 10)"] = df["test_score(out of 10)"].fillna(m)
df
```
