```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
%matplotlib inline
digits = load_digits()
```

```python
digits.data
```

```python
plt.gray()
for i in range(5):
    plt.matshow(digits.images[i])
```

```python
df = pd.DataFrame(digits.data)
df.head()
```

```python
df['target'] = digits.target
df.head()
```

```python
X = df.drop(['target'],axis='columns')
X.head()
```

```python
Y = df.target
Y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
len(X)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```

Changing parameter (n_estimator) in RandomForestClassifier to increse accuracy<br>
n_estimators is the n.o of trees in RandomForestClassifier

```python
model_n = RandomForestClassifier(n_estimators=56)
model_n.fit(X_train,Y_train)
model_n.score(X_test,Y_test)
```

```python
Y_predict = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn import metrics
cm = confusion_matrix(Y_test, Y_predict)
```

```python
import seaborn as sn
sn.heatmap(cm, annot=True)
plt.xlabel('Truth')
plt.ylabel('Predicted')
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df.head()
```

```python
iris_df['target'] = iris.target
iris_df.head()
```

```python
# iris_df['target_name'] = iris_df.target.apply(lambda x: iris.target_names[x])
# iris_df.head()
```

```python
iris_df.drop(['target_name'],axis='columns',inplace=True)
```

```python
X = iris_df.drop(['target'],axis='columns')
Y = iris_df.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
iris_model = RandomForestClassifier()
iris_model.fit(X_train,Y_train)
iris_model.score(X_test,Y_test)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sn
Y_predict = iris_model.predict(X_test)
iris_cm = confusion_matrix(Y_test, Y_predict)
sn.heatmap(iris_cm, annot=True)
plt.xlabel('Truth')
plt.ylabel('Predicted')
```

```python
iris_model_n = RandomForestClassifier(n_estimators=14)
iris_model_n.fit(X_train,Y_train)
iris_model_n.score(X_test,Y_test)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sn
Y_predict = iris_model_n.predict(X_test)
iris_cm = confusion_matrix(Y_test, Y_predict)
sn.heatmap(iris_cm, annot=True)
plt.xlabel('Truth')
plt.ylabel('Predicted')
```
